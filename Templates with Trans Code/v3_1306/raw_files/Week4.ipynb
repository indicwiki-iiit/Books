```python
import pickle 
import pandas as pd
import ast 
import pprint




d = {}
df = pickle.load(open('merged_pickle_v32.pkl','rb'))
for i in range(0,6500):
    if df.iloc[i]['goodreads_awards']!='':
        x = ast.literal_eval(df.iloc[i]['goodreads_awards'])
        for i in x:
            if i in d:
                d[i]+=1
            else:
                d[i]=1

l = list(d.items())
l = sorted(l,key = lambda x: x[1],reverse=True)
d = {}
for i in l:
    d[i[0]] = i[1]

x = list(d.items())
print(len(x))
print(pprint.pprint(x))
```

```python
from deeptranslit import DeepTranslit
trans = DeepTranslit('telugu')

transliterated_lis = []

for i in x:
    name_of_award  = i[0]
    freq_of_award  = i[1]
    trans_string   = name_of_award.replace("'","").replace("Award","Awaard")
    trans_of_award = trans.transliterate(trans_string)[0]['pred'].replace("\u200c","")
    transliterated_lis.append((name_of_award, freq_of_award, trans_of_award))
    
print(pprint.pprint(transliterated_lis))
```

```python
import pandas as pd

df = pd.read_csv("trans_award_names.csv")
c = 123
for i in transliterated_lis[123:151]:
    
    print("At #{} ".format(c),end=' ')
    name_of_award =  i[0]
    freq_of_award =  i[1]
    trans_of_award = i[2]
    
    print("{} {} {}".format(freq_of_award,name_of_award,trans_of_award))
    print("Enter change or 1 if the transliteration is right.")
    n = input()
    
    if n==1:
        pass
    else:
        trans_of_award = n
        
    df.at[c,'name_of_award']  = name_of_award
    df.at[c,'freq_of_award']  = freq_of_award
    df.at[c,'trans_of_award'] = trans_of_award
    
    c+=1
    
df.to_csv("trans_award_names.csv")
```
