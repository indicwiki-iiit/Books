### Code used for the analysis of number of words corresponding to each attribute in the article

```python
"""
@author: Pranav
"""

import pandas as pd
import numpy as np
import ast
import pickle
```

```python
df = pd.read_csv("ENG_TEL_DB.csv")
lis = ['BOOK_TITLE_TELUGU', 'SUBTITLE_TELUGU', 'BOOK_SUMMARY_TELUGU', 'AUTHORS_TELUGU', 'PUBLISHER_TELUGU', 'BOOK_EDITIONS_TELUGU', 'BOOK_SERIES_TELUGU', 'CHARACTERS_TELUGU', 'BOOK_AWARDS_TELUGU', 'AUTHOR_NAME_TELUGU', 'AUTHOR_NATIONALITY_TELUGU', 'AUTHOR_EDUCATION_TELUGU', 'AUTHOR_AWARDS_TELUGU', 'AUTHOR_NOTABLE_WORKS_TELUGU', 'AUTHOR_OTHER_WORKS_TELUGU', 'AUTHOR_GENRE_TELUGU', 'PUBLISHER_COLLECTION_TELUGU']
df = df.loc[0:6500,lis]
df.to_csv("dba1.csv")
```

```python
df = pd.read_csv("dba1.csv")
df2 = pd.read_csv("dba2.csv")
singular_lis = ['BOOK_TITLE_TELUGU','SUBTITLE_TELUGU','BOOK_SUMMARY_TELUGU','PUBLISHER_TELUGU','AUTHOR_NATIONALITY_TELUGU','AUTHOR_EDUCATION_TELUGU']
for j in singular_lis:
    print(j)
    for i in range(0,6500):
        print(i)
        e = df.iloc[i][j]
        if pd.isnull(e):
            df2.at[i,j] = 0
        else:
            x = e.split(" ")
            df2.at[i,j] = len(x)
    
df2.to_csv("dba2.csv")
```

```python
df = pd.read_csv("dba1.csv")
df2 = pd.read_csv("dba2.csv")

plural_lis = [
'BOOK_EDITIONS_TELUGU', 
'BOOK_SERIES_TELUGU', 
'CHARACTERS_TELUGU', 
'BOOK_AWARDS_TELUGU', 
'AUTHOR_AWARDS_TELUGU', 
'AUTHOR_NOTABLE_WORKS_TELUGU', 
'AUTHOR_OTHER_WORKS_TELUGU', 
'AUTHOR_GENRE_TELUGU', 
'PUBLISHER_COLLECTION_TELUGU']

for j in plural_lis:
    print(j)
    for i in range(0,6500):
        print(i)
        e = df.iloc[i][j]
        if pd.isnull(e) or e=='' or e=='[]':
            df2.at[i,j] = 0
        else:
            if e[0]!='[':
                df2.at[i,j] = 1
            else:
                try:
                    x = ast.literal_eval(e)
                    df2.at[i,j] = len(x)
                except:
                    if i==4298 or i==5321:
                        df2.at[i,j]= 0
                    else:
                        df2.at[i,j] = 5

df2.to_csv("dba3.csv")
```

```python
def amazonia(inp):
    if inp=='':
        return '0','0','0'
    else:
        k = list(inp.keys())[0]
        d = inp[k]

        rat = 0
        rac = 0
        dim = 0

        if 'Dimensions' in d:
            dim = 1
        if 'Ratings count' in d:
            rac = 1
        if 'Rating' in d:
            rat = 1

        return dim, rat, rac

df = pickle.load(open("merged_pickle_v37.pkl","rb"))
df2 = pd.read_csv("dba3.csv")
df2 = df2.astype({'DIMENSIONS':str,'AMAZON_RATING':str,'AMAZON_RATING_COUNT':str})
for i in range(0,6500):
    print(i)
    dim, rat, rac = amazonia(df.iloc[i]['amazon_book_specifications'])
    df2.at[i,'DIMENSIONS'] = str(dim)
    df2.at[i,'AMAZON_RATING'] = str(rat)
    df2.at[i,'AMAZON_RATING_COUNT'] = str(rac)

df2.to_csv("dba3.csv")
```

```python
df = pd.read_csv("dba3.csv")
for i in range(0,6500):
    x = df.iloc[i]['AUTHORS_TELUGU']
    try:
        temp = ast.literal_eval(x)
        c = 0
        for j in temp:
            y = j.split(" ")
            c += len(y)
        val = c
    except:
        print(i)
        val = 2

    df.at[i,'AUTHORS_TELUGU'] = val

df.to_csv("dba3.csv")
```

```python
lis = ['awards_as_category','amazon_formats_and_links']
df = pd.read_csv("dba6.csv")
df = df.astype({'awards_as_category':str,'amazon_formats_and_links':str})
df2 = pickle.load(open("merged_pickle_v37.pkl","rb"))

for j in lis:
    print(j)
    for i in range(0,6500):
        print(i)
        x = df2.iloc[i][j]
        if x=='':
            df.at[i,j] = 0
        else:
            if j=='amazon_formats_and_links':
                df.at[i,j] = len(x)
            elif j=='awards_as_category':
                c = 0
                for k in x:
                    c+= len(k.split(" "))
                df.at[i,j] = c

df.to_csv("dba7.csv")
```
