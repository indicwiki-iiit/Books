```python
# Function to be directly used to find the values failing ast.literal_eval

def ast_failure(df,ATT_NAME):
    lis = []
    for i in range(0,6500):
        x = df.iloc[i][ATT_NAME]
        if pd.isnull(x)==False:
            try:
                y = ast.literal_eval(x)
            except:
                print(i,x)
                lis.append(i)
    print(lis)
    print(len(lis))
```

### 1) Book Editions

#### 1.1 Identifying ast.literal_eval() failing values

```python
ast_failure(df,'BOOK_EDITIONS_TELUGU')
```

From the output, we see that values of 'BOOK_EDITIONS_TELUGU' and 'BOOK_EDITIONS_ENGLISH' at 1090, 2529, 3077, 3147, 4298, 4966, 5321 should be made null or np.nan.

```python
temp = [1090, 2529, 3077, 3147, 4298, 4966, 5321]
tlis = ['BOOK_EDITIONS_TELUGU','BOOK_EDITIONS_ENGLISH']

for j in tlis:
    for i in temp:
        df.at[i,j] = np.nan
```

#### 1.2 Correcting the value స్ట్. to make it సెయింట్ 

```python
number_of_replacements = 0

for i in range(0,6500):
    x = df.iloc[i]['BOOK_EDITIONS_TELUGU']
    if pd.isnull(x)==False:
        if 'స్ట్' in x:
            if 'స్ట్ మార్టిన్స్ ప్రెస్' in x or "స్ట్. మార్టిన్'స్ ప్రెస్" in x:
                df.at[i,'BOOK_EDITIONS_TELUGU'] = x.replace("స్ట్ మార్టిన్స్ ప్రెస్","సెయింట్ మార్టిన్స్ ప్రెస్").replace("స్ట్. మార్టిన్'స్ ప్రెస్","సెయింట్ మార్టిన్స్ ప్రెస్")
                number_of_replacements +=1
        else:
            pass
        
print("Number of replacements : {}".format(number_of_replacements))
```
