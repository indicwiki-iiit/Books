```python
# to fill missing values of author nationality, etc 
# usage of Google Search via Selenium

from selenium import webdriver
from time import sleep
from webdriver_manager.chrome import ChromeDriverManager
from datetime import date 
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time
import pandas as pd 


    
driver = webdriver.Chrome(ChromeDriverManager().install())
df = pd.read_csv("books_v5_spare.csv")
c = 0
d = 0

for i in range(90,95):
   
    a = df.iloc[i]['author1_name']
    if pd.isnull(a):
        print(i,"  ",a,"is NULL. Next iteration.")
        continue
    else:
        nation = df.iloc[i]['author1_nationality']
        if pd.isnull(nation)==True:

            print(i,"  ",a,"NATIONALITY IS NULL. Beginning to process....")
            c+=1
            authstring = a + " nationality"

            time.sleep(0.5)
            driver.get('https://www.google.com/')
            print("Opened Google")


            l = driver.find_element_by_xpath("//input[@class='gLFyf gsfi']")
            time.sleep(0.5)
            l.send_keys(authstring)
            l.send_keys(Keys.ENTER)
            time.sleep(4)

            try:
                l = driver.find_element_by_xpath("(//div[@class='Z0LcW XcVN5d'])[1]").text
                time.sleep(0.5)
                df.at[i,'author1_nationality'] = l
                d+=1
                print(l, "  ,done.")
            except NoSuchElementException:
                try:
                    l = driver.find_element_by_xpath("(//div[@class='LwV4sf'])[1]").text
                    time.sleep(0.5)
                    print(l)
                    df.at[i,'author1_nationality'] = l
                    d+=1
                    print(l, "  ,done.")
                except:
                    print("Element not found. Next iteration.")
                    continue
            
        else:
            print(i,"  ",a,"NATIONALITY IS #NOT# NULL. Next iteration....")

        df.to_csv("books_v5_spare.csv")

print(c)
print(d)
input('Press anything to quit') 
driver.quit() 
print("Finished") 
```
