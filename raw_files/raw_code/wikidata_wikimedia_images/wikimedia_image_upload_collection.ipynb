In the following cell, a list of files can be seen, starting with the name 'query'. These are csv files obtained after querying on Wikidata. However, all books do not belong to the same class, so the query had to be written for multiple classes named 'book', 'novel', 'literary work', 'written work', 'text', etc. because all of these classes on Wikidata contain details of books.

Each of these csv files contains just two attributes : 'item' ( which is a unique ID for each book/object on Wikidata starting with Q, for example Q588531 ) and 'itemLabel' containing the name of the book. 

**1A** : Books from all the below mentioned csv files have been added to the dictionary 'fin_lis' in the format {item:itemLabel} and it was found that 117551 records belong to this dictionary.

**1B** : Book names from our existing dataset, i.e. 'merged_pickle_v34.pkl' are queried against the book names which exist in the dictionary. The status of the book being found or not found in the dictionary is written to a new csv file, 'wiki_exists_modified.csv'. Along with the status, the value of the attribute 'item' is also written to the csv file which will act as a unique ID to search for it on Wikidata, in later stages of this code.

```python

#1A 


import pandas as pd
import pickle


lis = [
    'query_book.csv',
    'query_novel.csv',
    'query_creative_work.csv',
    'query_literary_work.csv',
    'query_novel.csv',
    'query_written_work.csv',
    'query_text.csv'
]

fin_lis = {}

for i in lis:
    df = pd.read_csv(i)
    x = len(df.item)
    for j in range(0,x):
        fin_lis.update({df.iloc[j]['item']:df.iloc[j]['itemLabel']})
    
```

```python
#2A, but the function is defined here and the main block in the next cell

def get_wikipedia_url_from_wikidata_id(wikidata_id, lang='en', debug=False):
    import requests
    from requests import utils

    url = (
        'https://www.wikidata.org/w/api.php'
        '?action=wbgetentities'
        '&props=sitelinks/urls'
        f'&ids={wikidata_id}'
        '&format=json')
    json_response = requests.get(url).json()
    if debug: print(wikidata_id, url, json_response) 

    entities = json_response.get('entities')    
    if entities:
        entity = entities.get(wikidata_id)
        if entity:
            sitelinks = entity.get('sitelinks')
            if sitelinks:
                if lang:
                    # filter only the specified language
                    sitelink = sitelinks.get(f'{lang}wiki')
                    if sitelink:
                        wiki_url = sitelink.get('url')
                        if wiki_url:
                            return requests.utils.unquote(wiki_url)
                else:
                    # return all of the urls
                    wiki_urls = {}
                    for key, sitelink in sitelinks.items():
                        wiki_url = sitelink.get('url')
                        if wiki_url:
                            wiki_urls[key] = requests.utils.unquote(wiki_url)
                    return wiki_urls
    return None   

get_wikipedia_url_from_wikidata_id('Q1317820', 'en')
```

```python
from selenium import webdriver
from time import sleep
from webdriver_manager.chrome import ChromeDriverManager
from datetime import date 
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time
import pandas as pd 
import pickle


driver = webdriver.Chrome(ChromeDriverManager().install())
c = 0

df = pd.read_csv("wiki_exists_final2.csv")
df = df.astype({'image':str})
for i in range(25,6500):
    print("At {}".format(i),end='  ')
    x = df.iloc[i]['wiki_link']
    
    if pd.isnull(x):
        print("NULL element.")
    else:
        print("Non Null element.",end='  ')
        driver.get(x)
        time.sleep(2)
        try:
            a = driver.find_element_by_class_name("infobox-image")
            b = a.find_element_by_tag_name("a")
            c = b.get_attribute("href")
            print("Image found.",end=' ')
            df.at[i,'image'] = c
            df.to_csv("wiki_exists_final2.csv")
            print("Appended to the DB")
        except:
            print("Image does not exist.")
        
```
